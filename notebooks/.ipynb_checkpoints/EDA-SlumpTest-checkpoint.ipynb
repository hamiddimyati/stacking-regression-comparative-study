```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import pandas_profiling
import seaborn as sb
```

```python
dir = "C:/Users/user/Documents/GitHub/II2202_Stacking-Regression-Comparative-Study/data"
```

```python
slump = pd.read_csv(dir + "/slump_test.data", header=0, encoding= 'unicode_escape')
```

```python
slump.info()
```

```python
slump.drop(columns=['No', 'SLUMP(cm)', 'FLOW(cm)'], inplace=True)
slump.rename(columns={'Fly ash': 'FlyAsh', 'Coarse Aggr.': 'CoarseAggr', 'Fine Aggr.': 'FineAggr', 'Compressive Strength (28-day)(Mpa)': 'CompressiveStr'}, inplace=True)
```

```python
pandas_profiling.ProfileReport(slump)
```

```python
slump.describe().transpose()
```

```python
plt.subplots(figsize=(12, 8))
ax = sb.boxplot(data=slump)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

```python
import itertools

cols = [i for i in slump.columns if i != 'CompressiveStr']

fig = plt.figure(figsize=(15, 20))

for i,j in itertools.zip_longest(cols, range(len(cols))):
    plt.subplot(4,2,j+1)
    ax = sb.distplot(slump[i],color='orange',rug=True)
    plt.axvline(slump[i].mean(),linestyle="dashed",label="mean", color='black')
    plt.legend()
    plt.title(i)
    plt.xlabel("")
```

```python
fig = plt.figure(figsize=(12, 8))
plt.axvline(slump.CompressiveStr.mean(),linestyle="dashed",label="mean", color='red')
sb.distplot(slump.CompressiveStr);
```

```python
from scipy.stats import skew
numerical_features = slump.select_dtypes(include=[np.number]).columns
categorical_features = slump.select_dtypes(include=[np.object]).columns
skew_values = skew(slump[numerical_features], nan_policy = 'omit')
dummy = pd.concat([pd.DataFrame(list(numerical_features), columns=['Features']), 
           pd.DataFrame(list(skew_values), columns=['Skewness degree'])], axis = 1)
dummy.sort_values(by = 'Skewness degree' , ascending = False)
```

```python
g = sb.PairGrid(slump)
g.map_upper(plt.scatter)
g.map_lower(sb.lineplot)
g.map_diag(sb.kdeplot, lw=3, legend=True);
```

```python
plt.subplots(figsize=(12, 6))
corr = slump.corr('spearman')

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True

ax = sb.heatmap(data=corr, cmap='magma', annot=True, mask=mask)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45);
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(14,11))

ax  = fig.gca(projection = "3d")

plot =  ax.scatter(slump["Cement"],
           slump["CompressiveStr"],
           slump["FlyAsh"],
           linewidth=1,edgecolor ="k",
           c=slump["Slag"],s=100,cmap="cool")

ax.set_xlabel("Cement")
ax.set_ylabel("Compressive Strength")
ax.set_zlabel("Fly Ash")

lab = fig.colorbar(plot,shrink=.5,aspect=5)
lab.set_label("Slag",fontsize = 15)

plt.title("3D plot for cement,compressive strength and fly ash",color="blue")
plt.show()
```
